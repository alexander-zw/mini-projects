# Using Binomial to Estimate Poisson

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Visualization for my section

```python
from scipy.special import comb
from scipy import math
from math import factorial
```

```python
def binom_prob(n, p, k):
    return comb(n, k) * p**k * (1 - p)**(n - k)

def poisson_prob(lamb, k):
    return lamb**k / factorial(k) * np.e**(-lamb)
```

```python
ns = [3, 10, 30, 100, 300]
limit = 7

plt.figure()
for n in ns:
    p = 2.5 / n
    ks = np.arange(0, n)[:limit]
    probs = [binom_prob(n, p, k) for k in ks][:limit]

    plt.bar(ks, probs, align='center', alpha=0.3)
    print('.', end='')

plt.xticks(ks, ks)
plt.ylabel('probability')
plt.title('binomial distribution')
plt.legend(ns)

plt.show()
```

```python
ns = [3, 10, 30, 100, 300]
limit = 7
lamb = 2.5

plt.figure()
for n in ns:
    p = lamb / n
    ks = np.arange(0, n)[:limit]
    probs = np.array([binom_prob(n, p, k) for k in ks])[:limit]

    plt.plot(ks, probs)
    print('.', end='')

ks = np.arange(0, limit)
probs = np.array([poisson_prob(lamb, k) for k in ks])

plt.plot(ks, probs)
print('.', end='')
    
plt.xticks(ks, ks)
plt.xlabel('$k$')
plt.ylabel('$P(X=k)$')
plt.title('binomial distribution vs poisson')
plt.legend(ns + ['$\infty$'])

plt.show()
```
